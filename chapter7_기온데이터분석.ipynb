```python
import csv
data = csv.reader(open('seoultemp.csv', encoding = 'cp949'))
next(data)
april = []
oct = []
for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '04':
     april.append(float(row[-1]))
    if row[0].split('-')[1] == '10':
     oct.append(float(row[-1]))

import matplotlib.pyplot as plt
plt.boxplot([april,oct])
plt.show()
```

```python
import csv
data = csv.reader(open('seoultemp.csv', encoding = 'cp949'))
next(data)
april = []
oct = []
for row in data:
  if row[-1] != '' and row[-2] !='' :
    if row[0].split('-')[1] == '04':
     april.append(float(row[-1]) - float(row[-2]))
    if row[0].split('-')[1] == '10':
     oct.append(float(row[-1]) - float(row[-2]))

import matplotlib.pyplot as plt
plt.boxplot([april,oct])
plt.show()
```

```python
#1년 중 가장 일교차가 큰 달은?
#리스트안에 리스트를 넣다.
import csv
data = csv.reader(open('seoultemp.csv', encoding = 'cp949'))
next(data)
x = []
for i in range(12) :
  x.append([])

for row in data:
  if row[-1] != '' and row[-2] != '' :
    x[int(row[0].split('-')[1]) - 1].append(float(row[-1]) - float(row[-2]))

import matplotlib.pyplot as plt
plt.boxplot(x)
plt.show()

```
